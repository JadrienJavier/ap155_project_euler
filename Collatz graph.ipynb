```python
import matplotlib.pyplot as plt

N_values = []
T_values = []

for i in range(1, 10001):
    j = i
    length = 1
    while j != 1:
        if j % 2 == 0:
            j = j // 2
        else:
            j = 3 * j + 1
        length = length + 1
    
    N_values.append(i)
    T_values.append(length)

plt.scatter(N_values, T_values)
plt.show()
```
